## Task 1: Prediction using Decision Tree Algorithm 

### Importing the libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Importing the Dataset 

```python
df = pd.read_csv("C:\Sample_\ML\Iris\Iris.csv")
```

```python
df.head()
```

```python
df.info()
```

### Splitting the data for Model Building  

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
X[:5]
```

```python
y = df['Species']
```

```python
y[:5]
```

### Model Building and training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=4)
```

```python
model.fit(X_train,y_train)
```

### Making the Predictions

```python
y_pred = model.predict(X_test)
```

```python
print(y_test[:5])
```

```python
print(y_pred[:5])
```

```python
Accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
Accuracy
```

### Let us visualize the model prediction results 

```python
from sklearn import tree
```

```python
fn = df.columns[1:5]
cn = df['Species'].unique().tolist()
model.fit(X,y)
fig, axes = plt.subplots(nrows=1,ncols=1,figsize=(10,10),dpi=200)
tree.plot_tree(model,
              feature_names=fn,
              class_names=cn,
              filled = True 
              )
```

### Predicting by giving the new value for it 

```python
X_new = [[7.1,4.5,2.4,1]]
```
